# WEb Scarping Using Python

# geting  covid-19 Data  from https://prsindia.org/ website using python

```python
import requests as r
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://prsindia.org/covid-19/cases'
res=r.get(url)
res
```

```python
data=BeautifulSoup(res.text,'html.parser')
h=[]
d=data.find_all('table')
r=d[0].find_all('tr')
for i in r[0]:
    h.append(i.text.strip())
print(h)
```

```python
r=d[0].find_all('tr')
x=[]
for i in r[1:]:
    t=i.find_all('td')
    y=[i.text for i in t]
    x.append(y)
    
for i in x:
    i[0]=i[0][:i[0].find('India')]
    i[1]=i[1][:i[1].find(' ')]
    i[2]=i[2][:i[2].find(' ')]
    i[3]=i[3][:i[3].find(' ')]
    i[4]=i[4][:i[4].find(' ')]
    
df=pd.DataFrame(x,columns=h)
df

    


    
    
```
